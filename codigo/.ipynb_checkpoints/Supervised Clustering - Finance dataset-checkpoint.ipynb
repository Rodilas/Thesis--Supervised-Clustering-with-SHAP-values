```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from  matplotlib import style
from datetime import datetime

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.cluster import DBSCAN
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV, cross_val_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.mixture import GaussianMixture

from umap import UMAP
import umap
import umap.plot
from sklearn.preprocessing import OneHotEncoder
from sklearn.copmose import make_column_transformer
```

```python
filename="C:\\Users\\rodri\\Downloads\\crx.data"
df = pd.read_csv(filename, sep=",", header=None, names=["A1", "A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10", "A11", "A12", "A13", "A14", "A15", "Target"])
df.replace("?", np.NaN, inplace=True)    
```

```python
df
```

```python
dfCatVar=pd.get_dummies(df[["A1","A4","A5", "A6", "A7", "A9", "A10", "A12", "A13"]])
dfContVar=df.drop(["A1","A4","A5", "A6", "A7", "A9", "A10", "A12", "A13"], axis=1)
dfs=[dfCatVar,dfContVar]
DfEncoded=pd.concat(dfs, axis=1)
```

```python
DfEncoded=DfEncoded.dropna()
```

```python
corrmat=DfEncoded.drop(["Target"], axis=1).corr()
top_corr_feats=corrmat.index
plt.figure(figsize=(30,30))
g= sns.heatmap(DfEncoded[top_corr_feats].corr(), annot=True, cmap="RdYlGn")
```

```python
raw = umap.UMAP(
  n_components=2, n_neighbors=100).fit_transform(DfEncoded.drop(["Target"], axis=1))
```

```python
rawdata = np.array(raw)
rawdata= pd.DataFrame(rawdata, columns=["x", "y"])
rawdata["Target"]= DfEncoded.Target


plt.scatter(rawdata[rawdata.Target=="+"]["x"], rawdata[rawdata.Target=="+"]["y"],color= "green", label = "+") 
plt.scatter(rawdata[rawdata.Target=="-"]["x"], rawdata[rawdata.Target=="-"]["y"],color= "red", label = "-") 
plt.legend()
```

```python
pca=PCA(n_components=2)
```

```python
pcaraw=pca.fit_transform(DfEncoded.drop(["Target"], axis=1))
```

```python
pcadata = np.array(pcaraw)
pcadata= pd.DataFrame(pcadata, columns=["x", "y"])
pcadata["Target"]= DfEncoded.Target

plt.figure(figsize=(10,10))
plt.scatter(pcadata[pcadata.Target=="+"]["x"], pcadata[pcadata.Target=="+"]["y"],color= "green", label = "+") 
plt.scatter(pcadata[pcadata.Target=="-"]["x"], pcadata[pcadata.Target=="-"]["y"],color= "red", label = "-") 
plt.legend()
```
